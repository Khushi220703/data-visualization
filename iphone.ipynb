```python
read_file = pd.read_csv("Order_details(masked).csv")
read_file
```

```python
read_file['Time'] = pd.to_datetime(read_file['Transaction Date'])
read_file
```

```python
read_file['Hour'] = (read_file['Time']).dt.hour
read_file
```

```python
time1 = read_file['Hour'].value_counts()
freq = []
for i in range(0,24):
   
    try:
        freq.append(time1[i])
    except:
        freq.append(0)
freq
```

```python
time = []
for i in range(0,24):
    time.append(i)
```

```python
plt.plot(time,freq)
plt.xlabel("Hour of buying")
plt.ylabel("Count of buying")
plt.title("Busiest hour")
plt.grid()
plt.show()
```
