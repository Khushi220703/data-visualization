```python
#import all required libraries.
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as pyp
```

```python
os.chdir("C:\\Users\\khush\\OneDrive\\Documents\\csv")
read_file = pd.read_csv("Audi.csv")
read_file.head()
```

```python
#checking for any null values.
print(read_file.count())
read_file.info()
```

```python
#no null values.
#counting and checking which model have most of sales.
sns.countplot(x="model", data=read_file)
pyp.xticks(rotation=90)
```

```python
#The model A3 have most of the sale, Q3 have secured rank two and A4 is third.
#While from RS5 to RS7 have negligible sale.

#checking transmission type.
x = read_file["model"]
y = read_file["transmission"]
sns.countplot(x = "transmission", data=read_file )
```

```python
#most of car is mannual.

sns.relplot(data=read_file, x='model', y='price', kind='line', errorbar=None,)
pyp.xticks(rotation=90)
```

```python
#model R8 have highest price.

sns.relplot(data=read_file, x='transmission', y='price', kind='line', errorbar=None,)
pyp.xticks(rotation=90)
```

```python
#automatic car have highest prices.

#mielage vs price
sns.relplot(data=read_file, x='mileage', y='price', kind='line', errorbar=None,)
pyp.xticks(rotation=90)
```

```python
# car with less mielage have higher price.
sns.relplot(data=read_file, x='engineSize', y='price', kind='line', errorbar=None,)
pyp.xticks(rotation=90)
```

```python
#larger the engine size higher the price.


sns.countplot(x ="fuelType", data=read_file)
```

```python
#car having type diesel have maximum sales.

#price vs fueltype.
sns.relplot(data=read_file, x='fuelType', y='price', kind='line', errorbar=None,)
pyp.xticks(rotation=90)
```
